```python
#importing standard libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#reading the data
train = pd.read_csv("C:/Users/HP/Documents/data science competitions/Mushroom dataset DSG-IITR/mushroom_train.csv")
test = pd.read_csv("C:/Users/HP/Documents/data science competitions/Mushroom dataset DSG-IITR/mushroom_test.csv")
```

Let's get dummies in one go.


```python
#creating a list of all the categorical variables. Will use it to get dummies in one go.
variables = []
for column in train.columns:
    variables.append(column)
variables.pop(12)
variables.pop(18)
variables
```

```python
#doing the same for test data
test_variables = variables.copy()
test_variables.pop(0)
test_variables
```

```python
#creating dummies
dummy_train = pd.get_dummies(train,columns=variables,drop_first=True)
```

```python
dummy_test = pd.get_dummies(test,columns=test_variables,drop_first=True)
```

## Model Building

```python
#making independent and target variables from training dataset
target = dummy_train['class_p']
training = dummy_train.drop('class_p',axis=1)
```

```python
#importing from sklearn
from sklearn.cross_validation import train_test_split
train_x,test_x,train_y,test_y = train_test_split(training,target,test_size=0.3)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

```python
#importing model from sklearn
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(train_x,train_y)
```

```python
predictions = model_rf.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y,predictions)
```
